# Health Risk Training (LSTM Model)
This notebook focuses on training the LSTM model using WESAD data for health risk assessment.

```python
# Load data logic from the relative path
%run ../data/Data_Management.ipynb

import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from tqdm.auto import tqdm

class HealthRiskModel:
    def __init__(self, input_shape):
        self.model = models.Sequential([
            layers.LSTM(64, input_shape=input_shape, return_sequences=True),
            layers.Dropout(0.2),
            layers.LSTM(32),
            layers.Dense(16, activation="relu"),
            layers.Dense(8, activation="softmax") # WESAD labels are 0-7
        ])
        self.model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])

loader = DataLoader()

# Using more subjects for better generalization
subjects = ["S2", "S3", "S4", "S5"]
X_list, y_list = [], []

for sub in tqdm(subjects, desc="WESAD Ingestion"):
    raw_data = loader.load_wesad_subject(sub)
    if raw_data:
        # Clean and filter labels
        df_clean = clean_wesad_data(raw_data)
        
        # Standardize features
        acc_data = df_clean[["acc_x", "acc_y", "acc_z"]].values
        labels = df_clean["label"].values
        
        # Normalize
        acc_norm = (acc_data - acc_data.mean(axis=0)) / (acc_data.std(axis=0) + 1e-7)
        
        # Create sequences (700 samples at 700Hz = 1 second of data)
        # Actually WESAD Chest ACC is 700Hz. Let's use 100 samples window.
        seq_len = 100
        X, y = [], []
        for i in range(0, len(acc_norm) - seq_len, seq_len):
            X.append(acc_norm[i:i + seq_len])
            # Most frequent label in window
            y.append(np.bincount(labels[i:i + seq_len]).argmax())
        
        X_list.append(np.array(X))
        y_list.append(np.array(y))

X_all = np.vstack(X_list)
y_all = np.concatenate(y_list)

X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=0.2, random_state=42)

print(f"Training shape: {X_train.shape}")
risk_model = HealthRiskModel(input_shape=(100, 3))
risk_model.model.fit(X_train, y_train, epochs=5, batch_size=64, validation_data=(X_test, y_test))

os.makedirs("../../models", exist_ok=True)
risk_model.model.save("../../models/health_risk_v1.h5")
print("Health Risk Model Saved Successfully.")
```
